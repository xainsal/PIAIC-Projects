```python
class SDMS():    
    def create_Student(name, fname, roll, Class, *Subjects):
        try:
            with open("students.txt" , 'a') as studnt:
                studnt.writelines(str(roll) + " " + name.title() + " " + fname.title() + " " + Class.title() + " " + str(Subjects))
                studnt.write("\n")
        except Exception as e:
            with open("students.txt" , 'w') as studnt:
                studnt.write("")
                print("Handle Error " + str(e))

    def get_StudentsByClass(Class):
        try:
            with open("Students.txt", 'r') as student:
                for line in student:
                    if Class in line:
                        print(line)
                    else:
                        continue
        except FileNotFoundError:
            print("Unable To Find Students.txt")

    def update_Student(roll, name):
        no = 0
        newline = []
        splited = []
        combined = ""
        count = 0
        new_name = input("Enter New Name: ").title()
        new_fname = input("Enter New Father Name: ").title()
        new_Class = input("Enter New Class: ").title()
        #opening File and saving it in a list
        try:
            with open("students.txt") as rf:
                for line in rf:
                    newline.append(line)
        except FileNotFoundError:
            print("unable to find 'students.txt'")
        else: 
            for x in newline:
                if roll in x and name in x:
                    splited = x.split(" ")
                    #print(splited)
                    if new_name != "": #and ("1","2","3","4","5","6","7","8","9","0") not in new_name: "will put a for loop here"
                        splited[1] = new_name
                    #elif ["1","2","3","4","5","6","7","8","9","0"] in new_name: "will put a for loop here"
                    #    print("Name can not Contain Numbers")
                    if new_fname != "": #and ["1","2","3","4","5","6","7","8","9","0"] not in new_fname: "will put a for loop here"
                        splited[2] = new_fname
                    #elif ["1","2","3","4","5","6","7","8","9","0"] in new_fname: "will put a for loop here"
                    #    print("Father Name can not Contain Numbers")
                    if new_Class != "":
                        splited[3] = new_Class
                    combined = " ".join(splited)
                    newline[no] = combined
                    #print(newline[no])
                    count += 1
                no += 1
            if count == 0:
                print(f'Unable to Find Roll No: {roll} or Name: {name} in File')
            #Writing To File
            with open("students.txt", "w") as wf:
                for line in newline:
                    wf.write(line)
        
    def delete_Student(roll, Class):
        no = 0
        count = 0
        newline = []
        try:
            with open("students.txt") as f:
                for line in f:
                    newline.append(line)
        except FileNotFoundError:
            print("Unable to Find 'student.txt'")
        else: 
            for x in newline:
                if roll in x and Class in x:
                    print(newline[no])
                    del newline[no]
                    count += 1
                    break
                no +=1
            if count == 0:
                print(f'Unable to Find Roll No: {roll} or Class: {Class} in File')
        with open("students.txt", "w") as wf:
            for line in newline:
                wf.write(line)
            
    def subjects_List():
        listSub = ["English","Urdu","Islamiat","Pakistan Studies","","",""]
        print("\nCompulsary Subjects Are Already Selected.")
        print(f'\tSubjects List:\n\t\t1. {listSub[0]}\n\t\t2. {listSub[1]}\n\t\t3. {listSub[2]}\n\t\t4. {listSub[3]}\n')
        x = 4
        while x < 7:
            tempSub = str(input("Enter " + str(x+1) +" Subject Name: ")).title()
            if tempSub in listSub:
                print("This Subject Already Exists")
                x -= 1
                tempSub = ""
            else:
                for no in ["1","2","3","4","5","6","7","8","9","0"]:
                    if no in tempSub:
                        print("Invalid Subject Name, Subject Name Can not Contain Numbers or Symbols.")
                        x -= 1
                        tempSub = ""
            if tempSub != "":
                listSub[x] = tempSub
            x += 1
        return listSub

    def programMenu():
        print("\n\n")
        print(f'[=====================================================]')
        print(f'[ Enter "1" To Create Student                         ]')
        print(f'[ Enter "2" To View Students of a Class               ]')
        print(f'[ Enter "3" To Update Student Information Enter       ]')
        print(f'[ Enter "4" To Delete a Student Enter                 ]')
        print(f'[ To Exit Enter "X"                                   ]')
        print(f'[=====================================================]')
        Choice = str(input("Please Input a Choice: "))
        print("\n\n")

        if Choice == '1':
            in_Name = str(input("Enter Student's Name: ")).title()
            in_FName = str(input("Enter Father's Name: ")).title()
            in_roll = input("Enter Roll No. of Student: ").title()
            in_class = str(input("Enter Class: ")).title()
            listSub = SDMS.subjects_List()
            SDMS.create_Student(in_Name, in_FName, in_roll, in_class, listSub)
            return 1
        elif Choice == '2':
            clss = str(input("Enter Class to View Students List: ")).title()
            SDMS.get_StudentsByClass(clss)
            return 1
        elif Choice == '3':
            in_roll = str(input("Enter Roll No. of Student: "))
            in_Name = str(input("Enter Student's Name: ")).title()
            SDMS.update_Student(in_roll, in_Name)
            return 1
        elif Choice == '4':
            in_roll = input("Enter Roll No. of Student: ")
            in_class = str(input("Enter Class: ")).title()
            SDMS.delete_Student(in_roll, in_class)
            return 1
        elif Choice == 'X' or Choice == "x":
            return 2
        else:
            print(f'{Choice} is a Incorrect Option')
            return 1
```
